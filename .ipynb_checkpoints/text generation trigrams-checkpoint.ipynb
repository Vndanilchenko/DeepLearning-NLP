```python
# импортируем библиотеки
import numpy as np
import pandas as pd
import sys
import nltk
# nltk.download('stopwords')
from nltk import ngrams
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.utils import np_utils
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam
```

```python
# загрузим обучающую выборку
file = pd.read_excel("../train_data.xlsx")
file.shape
```

```python
messages = file[file['intent']=='будет ли просрочка, если внести оплату в день списания платежа?']['message'].tolist()
print(len(messages))
```

```python
ngram = 3
def tokenize_words(input):
    # функция препроцессинга
    input = input.lower()
    trigrams = list(ngrams(input, ngram))
    filtered = []
    for i in range(len(trigrams)):
        filtered.append("".join(trigrams[i]))
#     return " ".join(filtered)
    return filtered + ["eof"]
```

```python
print(tokenize_words(messages[0])[:10])
print(tokenize_words(messages[0])[-10:])
```

```python
# препроцессинг
processed_inputs = []
for i in range(len(messages)):
    processed_inputs.append(tokenize_words(messages[i]))

processed_inputs[:5]
```

```python
# посчитаем среднюю и мин длину сообщения
mess_mean = int(np.mean([len(i) for i in processed_inputs]))
mess_min = int(np.min([len(i) for i in processed_inputs]))
mess_max = int(np.max([len(i) for i in processed_inputs]))
print("mean:",mess_mean, '\nmin:', mess_min,'\nmax:', mess_max)
```

```python
chars = set()
for i in processed_inputs:
    for j in i:
        chars.add(j)

print('элементов:{}'.format(len(chars)))
```

```python
char_to_num = dict((c, i) for i, c in enumerate(chars))

char_to_num
```

```python
input_len = 5
vocab_len = len(chars)
print ("Total number of characters:", input_len)
print ("Total vocab:", vocab_len)
```

```python
x_data = []
y_data = []
```

```python
# нарежем данные, разделив на последовательность элементов, по которой будем предсказывать таргет, также запишем сам таргет
for j in range(len(processed_inputs)):
    for i in range(0, len(processed_inputs[j]) - ngram):
        # запишем последовательность элементов, по которой будем предсказывать таргет
        in_seq = []
        for k in range(3):
            tmp_val = processed_inputs[j][i+k*ngram]
            in_seq.append(tmp_val)
#         print(in_seq)
        # будем выходить из цикла, если последовательность элементов менее ngram
        if len(in_seq)<ngram:
            break

        # таргет - элемент, который будем пытаться предсказать после поданной последовательности
        out_seq = processed_inputs[j][i+k*ngram+ngram]
        
#         print(out_seq)
            

        # преобразуем набор симоволов в индексы
        x_data.append([char_to_num[char] for char in in_seq])
        y_data.append(char_to_num[out_seq])
        
        # будем выходить из цикла, если достигли последнего токена eof
        if out_seq=='eof':
            break
```

```python
y = np_utils.to_categorical(y_data)
```

```python
y.shape
```

```python
# модель
model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', 
              optimizer=Adam(lr=0.0001))
filepath = "./model_weights_generated_trigrams_seq.hdf5"
model.load_weights(filepath)
checkpoint = ModelCheckpoint(filepath, 
                             monitor='loss', 
                             verbose=0, 
                             save_best_only=True, 
                             mode='min')
desired_callbacks = [checkpoint]
```

```python
model.fit(X, 
          y, 
          epochs=1000, 
          batch_size=64, 
          callbacks=desired_callbacks,
          verbose=1)
```

# генерация на обученной модели

```python
# подготовим обратный словарь индекс в элемент
num_to_char = dict((i, c) for i, c in enumerate(chars))
```

```python
# модель
model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', 
              optimizer=Adam(lr=0.0001))
filepath = "./model_weights_generated_trigrams_seq.hdf5"
model.load_weights(filepath)
```

```python
# выполним предсказание

def predict(pattern):
    x = np.reshape(pattern, (1, len(pattern), 1))
    x = x / float(vocab_len)
    prediction = model.predict(x, verbose=0)
    index = np.argmax(prediction)
    return index

# start = np.random.randint(0, len(x_data) - 1)
start = np.random.randint(0, len(x_data))
pattern = x_data[start]
print("случайная последовательность триграмм:")
print("\"", ''.join([num_to_char[value] for value in pattern]), "\"")
print("\nрезультат генерации:")
sys.stdout.write(''.join([num_to_char[value] for value in pattern]))
for i in range(1000):

    index = predict(pattern)
    result = num_to_char[index]
    
    if result=='eof':
        break
        
    sys.stdout.write(result)
        
    pattern.append(index)
    pattern = pattern[1:len(pattern)]
```
