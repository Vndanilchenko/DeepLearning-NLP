# DL based Text Recognition (OCR) using Tesseract

Optical Character Recognition(OCR) - извлечение текста с изображений

Данильченко Вадим

```python
from PIL import Image
import pytesseract
```

параметры:
- имя файла
- OCR language:язык, по умолчанию eng, русский указывается как rus, можно миксовать eng+rus, задается с помощью -l.
- OCR Engine Mode (oem): режим движка
    0. Legacy engine only.
    1. Neural nets LSTM engine only.
    2. Legacy + LSTM engines.
    3. Default, based on what is available.
- Page Segmentation Mode (psm): используется когда известна структура текста, по умолчанию стоит 3 (автовыбор)

```python
img1 = Image.open('../data/2tesseract4.png')
img1
```

```python
print(pytesseract.image_to_string(img1, config=('-l eng --oem 1 --psm 3')))
```

```python
img2 = Image.open('../data/2tesseract5.png')
img2
```

```python
print(pytesseract.image_to_string(img2, config=('-l eng+rus --oem 3 --psm 3')))
```
